```python
import os
import pandas as pd
import numpy as np
from itertools import chain, combinations
```

```python
files = os.listdir('csv/')
```

```python
files = sorted(list(filter(lambda x: '.csv' in x, files)))
```

```python
letters = ['A', 'B', 'C', 'D', 'E']
powerset = list(chain.from_iterable(combinations(letters, r) for r in range(len(letters)+1)))
letters_powerset = list(map(''.join, powerset))[1:]
columns = ['I'] + letters_powerset + ['E[nq]', 'E[w]', 'U']
signals = pd.DataFrame(columns=columns)
```

```python
lim = 2**5
for num in np.arange(lim):
    digits = [int(i) for i in format(num, '05b')]
    signals = signals.append(pd.Series(digits, letters), ignore_index=True)
```
